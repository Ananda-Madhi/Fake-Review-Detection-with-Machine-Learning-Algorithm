```python
import pandas as pd
import numpy as np
```

```python
file_path = r"C:\Users\ASUS\fr_dataset_stemmed.xlsx"
df = pd.read_excel(file_path)
```

```python
print("Data Awal:")
print(df.head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
import pandas as pd

print("TF-IDF Vectorizer...")
X_text = df['stemming'].astype(str)
y = df['label']

vectorizer = TfidfVectorizer()
X_vec = vectorizer.fit_transform(X_text)

print("Ukuran matriks fitur (dokumen, kata unik):", X_vec.shape)
print("-" * 40)

features = vectorizer.get_feature_names_out()
idf = vectorizer.idf_

rows = []
for i in range(X_vec.shape[0]):
    row_data = {}
    for idx in X_vec[i].indices:
        word = features[idx]
        tfidf_val = X_vec[i, idx]
        idf_val = idf[idx]
        tf_val = tfidf_val / idf_val
        row_data[word] = {"TF": tf_val, "IDF": idf_val, "TFIDF": tfidf_val}
    rows.append(row_data)

df_out = df.copy()
df_out["TF"] = [str({w: v["TF"] for w,v in r.items()}) for r in rows]
df_out["IDF"] = [str({w: v["IDF"] for w,v in r.items()}) for r in rows]
df_out["TFIDF"] = [str({w: v["TFIDF"] for w,v in r.items()}) for r in rows]

output_path = r"C:\Users\ASUS\fr_tfidf.xlsx"
df_out.to_excel(output_path, sheet_name="fr_tf_idf", index=False)

print(f"Hasil TF, IDF, TFIDF berhasil disimpan ke {output_path}")
```
