```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
file_path = r"C:\Users\ASUS\fr_dataset_stemmed.xlsx"
df = pd.read_excel(file_path)
```

```python
print("Data Awal:")
print(df.head())
```

```python

X = df['stemming'].astype(str)
vectorizer = CountVectorizer()
X_vec = vectorizer.fit_transform(X)

print("Model Bag of Words telah dibuat.")
print("Ukuran sparse matrix (dokumen, kata unik):", X_vec.shape)

print("\nContoh 20 kata pertama dari kosakata (fitur):")
print(vectorizer.get_feature_names_out()[:20])

X_array = X_vec.toarray()

features = vectorizer.get_feature_names_out()

count_str = []
for row in X_array:
    nonzero = [f"{features[i]}:{row[i]}" for i in row.nonzero()[0]]
    count_str.append(", ".join(nonzero))

df["countvectorizer"] = count_str

output_path = r"C:\Users\ASUS\fr_countvectorizer.xlsx"
with pd.ExcelWriter(output_path, engine="openpyxl") as writer:
    df.to_excel(writer, sheet_name="fr_countvectorizer", index=False)
```
